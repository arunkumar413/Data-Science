## Regression plots

### using lmplot( ) method

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue= 'time')
```

```python
# with markers  list of other markers: https://matplotlib.org/api/markers_api.html

sns.lmplot(x='total_bill',y='tip',data=tips, hue= 'time', markers= ["x","o"], scatter_kws={"s": 100})
```

```python
sns.lmplot(x="total_bill", y= 'tip',data=tips, col= 'sex') # insteas of hue you can use the col parameter to plot seperately
```

```python
sns.lmplot(x="total_bill", y= 'tip',data=tips, col= 'sex',row ='time')  # with both row and column parameter
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')  # with col and hue parameter
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='day',hue='sex')  # with row and hue parameter
```

```python
#with custom aspect ratio and size

sns.lmplot(x="total_bill",y="tip",data=tips, col= 'sex', hue="time", aspect=0.7,size=8)
```
