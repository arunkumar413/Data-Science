# Distribution plots

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot


```python
# import modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")  ## import tips data set
```

```python
tips.head()                   ## check the dataframe structure
```

```python
##distribution plot with KDE. This the equivalanet of the pandas histogram plot   

sns.distplot(tips["total_bill"])
```

```python
##distribution plot without KDE and custom bins

sns.distplot(tips["total_bill"],kde=False,bins=30)
```

# Joint plot

##  types: 

* scatter
* reg
* kde
* hex

```python
# Joint plot with kind = scatter
sns.jointplot(x= "total_bill", y= "tip",data= tips,kind="scatter")
```

```python
# joint plot with kind = hex
sns.jointplot(x="total_bill",y="tip",data= tips, kind= "hex")
```

```python
sns.jointplot(x="total_bill", y='tip',data = tips, kind= "reg")   ## joint plot with regression type
```

## Pair plot

```python
## Pairplot

sns.pairplot(tips)
```

```python
#pairplot with hue

sns.pairplot(tips, hue= 'sex')
```

## Rug plot

```python
## Rug plot

sns.rugplot(tips["total_bill"])
```

## kde plot

```python
#kde plot
sns.kdeplot(tips["total_bill"])
sns.rugplot(tips["total_bill"])
```

```python
sns.kdeplot(tips["tip"])
sns.rugplot(tips["tip"])
```
