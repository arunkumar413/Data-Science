## Categorical plots

* factorplot
* boxplot
* violinplot
* striplot
* swarmplot
* barplot
* countplot


```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

### barplot and countplot

```python
sns.barplot(x="sex", y="total_bill",data= tips)
```

```python
sns.barplot(x="sex",y="total_bill",data= tips, hue="time")  # with hue
```

```python
sns.countplot(x='sex',data= tips)    ## Just plots the number of occurences of the parameter
```

```python
sns.countplot(x='time',data= tips)    ## Just plots the number of occurences of the parameter
```

```python
sns.countplot(x='sex',data= tips, hue="smoker")    ## we can use hue as well in countplot
```

## boxplot and violinplot

```python
sns.boxplot(x="sex", y='total_bill',data= tips)
```

```python
sns.boxplot(x="smoker", y='total_bill',data= tips)
```

```python
sns.boxplot(x="day", y='total_bill',data= tips, hue= "time")  # with hue
```

## violinplot

```python
sns.violinplot(x="sex", y="total_bill",data= tips)
```

```python
sns.violinplot(x="day", y="total_bill",data= tips)
```

```python
sns.violinplot(x="time", y="total_bill",data= tips)
```

```python
## with hue and split = true. Makes comparision easier
sns.violinplot(x="day", y="total_bill",data= tips, hue="sex", split=True)
```

```python
sns.violinplot(x="day", y="total_bill",data= tips, hue="sex")   ## with hue and split = false
```

## stripplot and swarmplot

```python
sns.stripplot(x='day',y="total_bill",data= tips)
```

```python
sns.stripplot(x='day',y='total_bill', data= tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data= tips, jitter=True, hue='time')
```

## swarmplot

```python
sns.swarmplot(x='day', y = "total_bill",data=tips)
```

```python
sns.swarmplot(x='day', y = "total_bill",data=tips, hue="time",palette="Set1")
```

```python
sns.swarmplot(x='day', y = "total_bill",data=tips, hue="time",palette="Set1", split=True)
```

## combining swarm and violinplot

```python
sns.violinplot(x="tip", y="time", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="time", data=tips,color='black',size=3)
```

## factor plot

Factor plot is a general form of categorical plot. The plot type can specified by the kind paramenter

kind can take ``point``, ``bar``, ``count``, ``box``, ``violin``, ``strip``


```python
sns.factorplot(x="time", y= "total_bill", data= tips, kind='bar')
```

```python
sns.factorplot(x="time", y= "total_bill", data= tips, kind='box')
```

```python
sns.factorplot(x="time", y= "total_bill", data= tips, kind='strip')
```
