## Matrix Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flights = sns.load_dataset("flights")  ## load flights data set
tips    = sns.load_dataset("tips")     ## load tips dataset
```

```python
flights.head()
```

```python
tips.head()
```

### convert these data frames into matrix form

```python
tips_corr = tips.corr()   ## convert the dataframe into matri format uisng corr() method
tips_corr.head()
```

```python
sns.heatmap(tips_corr)
```

```python
sns.heatmap(tips_corr, annot=True)  # label the boxes
```

## Using pivot table

```python
flights.head()
```

```python
flights_table = flights.pivot_table(index= 'year', columns='month',values='passengers')  ## create a pivot table
flights_table.head()
```

```python
sns.heatmap(flights_table)   ## create a heat map
```

```python
plt.figure(figsize=(15,10))  ## custom size
sns.heatmap(flights_table, annot=True, fmt='g')
## use the parameter fmt = 'g' otherwise annotation would show up in exponential format
```

```python
plt.figure(figsize=(15,10))  ## custom size
sns.heatmap(flights_table, annot=True, fmt='g',cmap='Blues')   # blue color map
## use the parameter fmt = 'g' otherwise annotation would show up in exponential format
```

## Cluster Map

```python
sns.clustermap(tips_corr,cmap="Blues")
```

```python
sns.clustermap(flights_table,cmap="Reds")
```
