# Filtering data in pandas

```python
import pandas as pd
import numpy as np
```

```python
# Create a dataframe of 1000 rows with column names A,B,C,D,E

df= pd.DataFrame(np.random.randint(1,100, size=(1000,5)),columns=list("ABCDE"))
df.head()
```

#### Create a few filters


```python
f1 = df['A']>50   #select the rows where the  Column 'A' values are are more than 50
f2= df['C']==25   #select the rows where the  Column 'C' values are equal to 25
f3= df['D']>20    #select the rows where the  Column 'D' values are greater than 20 
f4 = df['B']<20    #select the rows where the  Column 'B' values are lesser than 20
```

#### Apply the filter on the dataframe

```python
df[f1]   
```

```python
df[f2]   
```

#### Apply multiple filters using  '&'

```python
df[f3 & f2]   
```

#### You can also comibine AND, OR in the filters 

```python
df[(f1 | f4)& f2]  
```
