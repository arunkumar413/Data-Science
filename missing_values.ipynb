### Missing Values

One of the important aspect of Data Analysis is accounting or analysing the missing values in the Data. Luckily it's easier in Pandas to check or identify the missing values. In Pandas missing values are represented as "NaN"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Create a dataframe of 100 rows and 5 columns
df = pd.DataFrame(np.random.randint(50,100, size=(100,5)),columns=['A','B','C','D','E'])
```

```python
#create a few NaNs in the Dataframe

df.iloc[5:10,0]=  np.nan
df.iloc[30:60,1]= np.nan
df.iloc[70:76,2]= np.nan
df.iloc[85:95,3]= np.nan
```

#### We'll use isnull() method to check if each value is a NaN. 
#### isnull() method returns True if the value is NaN and False if the value is not NaN.


```python
df.isnull()
```

#### Display the NaN count for each column by applying sum() function

```python
df.isnull().sum()
```

#### Visualize NaNs for each column by applying the plot function

```python
df.isnull().sum().plot(kind="barh")
```

#### Display the missing values in descending order by using sort_vlaues() function

```python
df.isnull().sum().sort_values().plot(kind='barh')
```

#### Another way to visualize the missing values using seaborns barplot() function

```python
cols = df.columns    # list holds the column names
missing = []         # list holds the missing value count
for col in cols:
    i = df[col].isnull().sum()
    missing.append(i)
    
sns.barplot(x=cols,y=missing)    
    
```

#### As you may frequently want to analyse the missing values, you can convert these small snippets into a function 

```python
def missing_values(df):
    cols = df.columns    # Array holds the column names
    missing = []         # Array holds the missing value count
    for col in cols:
        i = df[col].isnull().sum()
        missing.append(i)
    
    print(df.isnull().sum())
    sns.barplot(x=cols,y=missing)  
    
```

```python
missing_values(df)  #call the function by passing dataframe as the parameter
```
